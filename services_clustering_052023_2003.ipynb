```python
# Import dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.cluster import KMeans 
```

```python
# Read in the CSV file as a Pandas DataFrame
mypath = Path('../Resources/service_ratings.csv')
df = pd.read_csv(mypath)

# Review the DataFrame
df.head()
```

```python
# Scatter plot of data
X = df['mobile_app_rating']
y = df['personal_banker_rating']
data = list(zip(X,y))

plt.figure(figsize=(20,10))
plt.scatter(X,y)
plt.show()
```

## Run the k-means model with 2 clusters

```python
# Start by importing the K-means algorithm
# already imported
```

```python
# Create and initialize the K-means model instance for 2 clusters
kmeans = KMeans(n_clusters=2, random_state=1)

# Print the model
print(kmeans)
```

```python
# Fit the data to the instance of the model
kmeans.fit(data)
```

```python
# Make predictions about the data clusters using the trained model
predictions_1 = kmeans.predict(data)

# Print the predictions
print(predictions_1)
```

```python
# Add a column to the DataFrame that contains the customer_ratings information
df['customer_ratings'] = predictions_1

# Review the DataFrame
df.head()
```

```python
# Plot the datapoints based on the customer rating
plt.figure(figsize=(20,10))
plt.scatter(X,y,c=df['customer_ratings'])
plt.show()
```
